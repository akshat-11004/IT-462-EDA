# Using missingno for Visulizing the missing data

```python
msno.bar(df)
```







**Analysis of the Generated Bar Chart:**



*   The majority of the columns, such as FL_DATE, AIRLINE, ORIGIN_CITY, ARR_DELAY, and many others, have complete data (non-missing), indicated by the bars reaching 1.0.
*   
Columns like CANCELLATION_CODE show very High missing values.


*   Columns toward the end, such as DELAY_DUE_TO_CARRIER, DELAY_DUE_WEATHER, and DELAY_DUE_LATE_AIRCRAFT, have a significant amount of missing data, as their bars are much shorter, indicating only about 20% of the data is available in these columns.










```python
msno.matrix(df)
```




**Analysis of the Generated Matrix Plot:**


*   Most columns like FL_DATE, AIRLINE, AIRLINE_DOT, ORIGIN_CITY, and ARR_DELAY have complete data, indicated by the continuous black lines with no gaps.

*   Some columns, such as TAXI_OUT, WHEELS_OFF, and TAXI_IN, have scattered missing values throughout the dataset, as seen from the white lines interspersed among black ones.
*   Columns towards the right, like DELAY_DUE_CARRIER, DELAY_DUE_WEATHER, and DELAY_DUE_LATE_AIRCRAFT, have significant portions of missing data, indicated by frequent white gaps, especially towards the bottom. This suggests that these features are only available for a subset of the dataset.







```python
msno.heatmap(df)
```




**Analysis of the Generated Heatmap:**




*   There is a strong correlation (> 1) between missing values in the CANCELLATION_CODE column and several other columns such as DEP_DELAY,
 ARR_DELAY, and AIR_TIME, indicating that missing data in these columns is highly dependent on the presence of missing values in CANCELLATION_CODE. This might suggest that flights with missing cancellation codes also tend to have delays or other issues, requiring further exploration.

*   Most other columns, like DEP_TIME, TAXI_OUT, and ARR_TIME, show correlations close to zero, suggesting that missing values in one feature do not strongly predict missingness in the others. For instance, DEP_TIME and DEP_DELAY have little correlation with missing values in the DELAY_DUE_WEATHER and DELAY_DUE_LATE_AIRCRAFT columns.

*   There are negative correlations (less than -1) observed between CANCELLATION_CODE and features like DELAY_DUE_CARRIER and DELAY_DUE_LATE_AIRCRAFT, indicating that when one of these columns has missing data, the other is more likely to have complete data.


*   The presence of both strong and weak correlations points to different patterns of missingness across the dataset. Columns with high correlations in missing values may benefit from joint imputation, while columns with weaker correlations may require different approaches, such as removal or independent imputation.





```python
msno.dendrogram(df)
```



**Analysis of the Generated Dendrogram:**



*   The dendrogram groups columns based on similarities in their missing values. The closer two columns are on the dendrogram, the more similar their missing data patterns are. In this case, columns like "DELAY_DUE_WEATHER", "DELAY_DUE_CARRIER", and "DELAY_DUE_LATE_AIRCRAFT" cluster together, indicating that they share similar missing value patterns.

*   one cluster includes columns like "WHEELS_OFF", "TAXI_OUT", and "DEP_DELAY". This suggests that these columns tend to have similar missing data patterns, meaning they often have missing values at the same times or under the same conditions.

*   Columns that are connected at the same level (like "DELAY_DUE_WEATHER" and "DELAY_DUE_CARRIER") are highly similar in terms of missing data.
